```python
import pandas as pd
import numpy as np

np.random.seed(50)

#Crear dataset de inventario

datos_inventrio = {
    'CondigoProducto':[f'PROD{str(i).zfill(3)}'for i in range(1, 81)],
    'NombreProducto': np.random.choice(['Camiseta', 'Pantalón', 'Vestido', 'Chaqueta',
    'Zapatos', 'Falda'], 80),
    'Categoria': np.random.choice(['Hombre', 'Mujer', 'Niño'], 80),
    'Talla': np.random.choice(['S', 'M', 'L', 'XL'], 80),
    'PrecioCompra': np.random.randint(20000, 100000, 80),
    'PrecioVenta': np.random.randint(40000, 200000, 80),
    'Stock': np.random.randint(0, 50, 80),
    'UnidadesVendidas': np.random.randint(0, 100, 80),
    'Proveedor': np.random.choice(['Proveedor_A', 'Proveedor_B', 'Proveedor_C'], 80)
}

df_inventario = pd.DataFrame(datos_inventrio)

#introducir algunos valores nulos

df_inventario.loc[10:12,'Stock'] = np.nan
df_inventario.loc[25, 'UnidadesVendidas'] = np.nan

#Guardar dataset
df_inventario.to_csv('inventario_tienda.csv', index=False)
print("Dataset creado: inventario_tienda.csv")
print(df_inventario.head(10))
```

```python
# 1. Importar librerías
import pandas as pd

# 2. Cargar el archivo CSV
df = pd.read_csv('inventario_tienda.csv')

# 3. Mostrar información general del DataFrame
print("=== Información general del DataFrame ===")
print(df.info())
print("\nPrimeras filas del dataset:")
print(df.head())

# 4. Identificar y contar valores nulos
print("\n=== Conteo de valores nulos por columna ===")
print(df.isnull().sum())

# 5. Rellenar valores nulos en 'Stock' con 0
df['Stock'].fillna(0, inplace=True)

# 6. Rellenar valores nulos en 'UnidadesVendidas' con la mediana
mediana_unidades = df['UnidadesVendidas'].median()
df['UnidadesVendidas'].fillna(mediana_unidades, inplace=True)

# 7. Verificar que no queden valores nulos
print("\n=== Verificación de valores nulos después de la limpieza ===")
print(df.isnull().sum())

# 8. Mostrar el resultado final
print("\nDataFrame limpio - primeras filas:")
print(df.head())

# 9. Guardar el nuevo archivo limpio
df.to_csv('inventario_tienda_limpio.csv', index=False)
print("\nArchivo limpio guardado como 'inventario_tienda_limpio.csv'")
```

```python
#CALCULOS Y TRANSFORMACIONES
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('inventario_tienda_limpio.csv')

# 1. Calcular Ganancia
df['Ganancia'] = df['PrecioVenta'] - df['PrecioCompra']

# 2. Calcular MargenPorcentaje
df['MargenPorcentaje'] = (df['Ganancia'] / df['PrecioCompra']) * 100

# 3. Calcular ValorInventario
df['ValorInventario'] = df['Stock'] * df['PrecioVenta']

# 4. Calcular IngresosTotales
df['IngresosTotales'] = df['UnidadesVendidas'] * df['PrecioVenta']

# 5. Clasificar productos según stock
def clasificacion_stock(stock):
    if stock == 0:
        return 'Agotado'
    elif 1 <= stock <= 10:
        return 'Bajo'
    elif 11 <= stock <= 30:
        return 'Normal'
    elif stock > 30:
        return 'Alto'

df['ClasificacionStock'] = df['Stock'].apply(clasificacion_stock)

# Mostrar primeras filas del DataFrame transformado
print(df.head())


# Guardar el DataFrame como archivo CSV
df.to_csv('inventario_tienda_transformado.csv', index=False)
print("Archivo guardado como 'inventario_tienda_transformado.csv'")
```

```python
# === ANÁLISIS DE DATOS ===
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('inventario_tienda_transformado.csv')

# 1. Valor total del inventario
valor_total_inventario = df['ValorInventario'].sum()
print('\nValor total del inventario:', valor_total_inventario)

# 2. Top 5 productos más vendidos
top_5_vendidos = df.sort_values('UnidadesVendidas', ascending=False).head(5)
print('\nTop 5 productos más vendidos:\n', top_5_vendidos[['CondigoProducto', 'NombreProducto', 'UnidadesVendidas']])

# 3. Top 5 productos mayor margen
top_5_margen = df.sort_values('MargenPorcentaje', ascending=False).head(5)
print('\nTop 5 productos con mayor margen de ganancia:\n', top_5_margen[['CondigoProducto', 'NombreProducto', 'MargenPorcentaje']])

# 4. Ingresos totales por categoría
ingresos_por_categoria = df.groupby('Categoria')['IngresosTotales'].sum()
print('\nIngresos totales por categoría:\n', ingresos_por_categoria)

# 5. Stock promedio por proveedor
stock_promedio_proveedor = df.groupby('Proveedor')['Stock'].mean()
print('\nStock promedio por proveedor:\n', stock_promedio_proveedor)

# 6. Contar productos agotados
num_agotados = (df['ClasificacionStock'] == 'Agotado').sum()
print('\nProductos agotados:', num_agotados)

# 7. Proveedor con mejor margen
proveedor_mejor_margen = df.groupby('Proveedor')['MargenPorcentaje'].mean().idxmax()
print('\nProveedor con mejor margen promedio:', proveedor_mejor_margen)

# Guardar el DataFrame como archivo CSV
df.to_csv('inventario_tienda_analisis.csv', index=False)
print("Archivo guardado como 'inventario_analisis.csv'")
```

```python
import pandas as pd

# 1. Cargar el dataset transformado
df = pd.read_csv('inventario_tienda_transformado.csv')


# === 1. Productos agotados que hayan vendido más de 20 unidades ===
reabastecer_urgente = df[(df['ClasificacionStock'] == 'Agotado') & (df['UnidadesVendidas'] > 20)]
print("\n Productos AGOTADOS con más de 20 unidades vendidas (reabastecimiento urgente):")
print(reabastecer_urgente[['CondigoProducto', 'NombreProducto', 'Stock', 'UnidadesVendidas', 'Categoria', 'Proveedor']])

# === 2. Productos con stock alto (>30) y pocas ventas (<10) ===
candidatos_descuento = df[(df['Stock'] > 30) & (df['UnidadesVendidas'] < 10)]
print("\n Productos con STOCK ALTO y pocas ventas (candidatos a descuento):")
print(candidatos_descuento[['CondigoProducto', 'NombreProducto', 'Stock', 'UnidadesVendidas', 'Categoria', 'Proveedor']])

# === 3. Productos con margen mayor a 50% ===
margen_alto = df[df['MargenPorcentaje'] > 50]
print("\n Productos con MARGEN mayor al 50%:")
print(margen_alto[['CondigoProducto', 'NombreProducto', 'MargenPorcentaje', 'Categoria', 'Proveedor']])

# === 4. Reporte de productos por categoría y estado de stock ===
reporte_categoria_stock = df.groupby(['Categoria', 'ClasificacionStock']).size().reset_index(name='CantidadProductos')
print("\n Reporte de productos por CATEGORÍA y NIVEL DE STOCK:")
print(reporte_categoria_stock)

# === (Opcional) Guardar los resultados en archivos CSV individuales ===
reabastecer_urgente.to_csv('reabastecer_urgente.csv', index=False)
candidatos_descuento.to_csv('candidatos_descuento.csv', index=False)
margen_alto.to_csv('margen_alto.csv', index=False)
reporte_categoria_stock.to_csv('reporte_categoria_stock.csv', index=False)

print("\n Archivos de reportes generados correctamente:")
print("- reabastecer_urgente.csv")
print("- candidatos_descuento.csv")
print("- margen_alto.csv")
print("- reporte_categoria_stock.csv")
```

```python
import pandas as pd

# 1. Cargar el dataset transformado
df = pd.read_csv('inventario_tienda_transformado.csv')

# === 1. Exportar DataFrame completo procesado ===
df.to_csv('inventario_procesado.csv', index=False)
print(" Archivo exportado: inventario_procesado.csv")

# === 2. Exportar productos para reabastecer ===
# (Agotados con más de 20 unidades vendidas)
productos_reabastecer = df[(df['ClasificacionStock'] == 'Agotado') & (df['UnidadesVendidas'] > 20)]
productos_reabastecer.to_csv('productos_reabastecer.csv', index=False)
print(" Archivo exportado: productos_reabastecer.csv")

# === 3. Exportar productos para descuento ===
# (Stock alto >30 y pocas ventas <10)
productos_descuento = df[(df['Stock'] > 30) & (df['UnidadesVendidas'] < 10)]
productos_descuento.to_csv('productos_descuento.csv', index=False)
print(" Archivo exportado: productos_descuento.csv")

# === 4. Exportar resumen por proveedor ===
# Incluye total de productos, stock promedio, margen promedio e ingresos totales
resumen_proveedores = df.groupby('Proveedor').agg({
    'CondigoProducto': 'count',
    'Stock': 'mean',
    'MargenPorcentaje': 'mean',
    'IngresosTotales': 'sum'
}).reset_index()

# Cambiar nombre de columnas para que el reporte quede más claro
resumen_proveedores.columns = [
    'Proveedor',
    'CantidadProductos',
    'StockPromedio',
    'MargenPromedio(%)',
    'IngresosTotales'
]

# Exportar
resumen_proveedores.to_csv('resumen_proveedores.csv', index=False)
print(" Archivo exportado: resumen_proveedores.csv")

print("\n Exportación completada exitosamente.")
```
